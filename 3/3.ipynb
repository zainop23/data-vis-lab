```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('gapminder.csv')

px.line(df.query("continent=='Oceania'"), x='year', y='lifeExp', color='country', title='Life Exp year-wise', markers='country').show()
px.bar(df.query("year==2007 and continent=='Europe'"),x='country',y='gdpPercap',color='gdpPercap',title='Europe 2007 gdp per Capita').show()
px.scatter(df.query("year==2007"),x='gdpPercap',y='lifeExp',color='continent',size='pop',size_max=60, log_x=True).show()
```

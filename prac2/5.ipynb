```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('earthquakes-23k.csv')
df = df.dropna(subset=['Latitude', 'Longitude', 'Magnitude'])

fig = px.scatter_geo(
    df,
    lat='Latitude',
    lon='Longitude',
    size='Magnitude',
    color='Magnitude',
    hover_name='Magnitude',
    projection='natural earth',
    title='Earthquake map'
)

fig.show()
```

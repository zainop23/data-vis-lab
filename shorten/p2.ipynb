2. Write a program to create pair plots and regression plots using Seaborn, to study variate relationships and trends.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = pd.read_csv("../2/iris.csv")
```

```python
corr = iris.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
```

```python
sns.pairplot(iris,hue="species")
```

```python
sns.regplot(data=iris,x="sepal_length",y="petal_length",order=3)
```

```python
print(f"Pearson Correlation: {iris["sepal_length"].corr(iris["petal_length"]):.4f}")
```

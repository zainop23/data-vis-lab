Write a program to perform time series visualization using Pandas and Matplotlib,
showing trends, seasonal patterns, and moving averages.
Given the "Airline Passengers" dataset (1949–1960), write a Python program to:
- Pre-process the data by converting date strings into a DatetimeIndex.
- Calculate a 12-month Rolling Mean to smooth out noise and visualize the growth trend.
- Generate a Seasonal Boxplot to identify recurring monthly fluctuations (Seasonality).

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
```

```python
df = pd.read_csv("../1/file.csv")
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month',inplace=True)
```

```python
df['12-MA'] = df['Passengers'].rolling(window=12).mean()
```

```python
plt.plot(df)
plt.title("Moving Average Trend")
```

```python
df['Month_name'] = df.index.month_name()
sns.boxplot(data=df,x="Month_name",y="Passengers",hue="Month_name")
plt.xticks(rotation=45)
```

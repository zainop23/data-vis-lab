4. Write a program to design an interactive dashboard using Plotly, incorporating multiple charts, filters, and dropdown controls.

```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html
from dash.dependencies import Input, Output
```

```python
df = pd.read_csv("../4/tips.csv")
df
```

```python
app = Dash(__name__)


app.layout = html.Div([
    dcc.Dropdown(id="day", options=df["day"].unique(), value="Sun"),
    dcc.Dropdown(id="time", options=df["time"].unique(), value="Dinner"),
    *[dcc.Graph(id=str(i)) for i in range(4)],
])

@app.callback(
    [Output(str(i), "figure") for i in range(4)],
    [Input("day", "value"), Input("time", "value")],
)
def update(d, t):
    f = df.query("day == @d and time == @t")
    return (
        px.scatter(f, x="total_bill", y="tip", color="sex"),
        px.bar(f, x="sex", y="tip", color="smoker"),
        px.histogram(f, x="total_bill", color="sex"),
        px.box(f, x="sex", y="tip"),
    )
```

```python
app.run(port=6767)
```
